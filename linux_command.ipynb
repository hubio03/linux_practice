```python
# LiSt 의 ls
# 파일 및 디렉토리 목록 표시

# ls -l: 상세 정보를 롱 포멧으로 표시
# ls -alF: ls -a, ls -l, ls -F를 한번에 보여줌

# alias 로 ll == ls -l 의 기능을 함

!ls
```

```python
# Change Directory 의 cd
# 디렉토리 이동

# cd . 은 현재 디렉토리 명령어 자체로 사용하기보단 지금 여기에 파일 저장 하려하거나 할 때 사용
# cd .. 은 상위 디렉토리로 이동한다는 뜻

!cd test
```

```python
# Print Working Directory 의 pwd
# 현재 위치한 디렉토리 경로 출력

!pwd
```

```python
# ReMove 의 rm
# 파일이나 디렉토리를 삭제

# rm [파일명]: 파일 삭제
# rm -r [디렉토리명]: 디렉토리와 그 안의 내용을 모두 삭제 (Recursive)
# rm -f [파일명]: 강제로 삭제 (Force, 확인 메시지 없음)

# 리눅스에는 '휴지통' 개념이 기본적으로 없으므로, rm으로 지운 파일은 복구하기 매우 어렵다.

!rm
```

```python
# CoPy 의 cp
# 파일이나 디렉토리를 복사

# cp [원본] [대상]: 원본을 대상 이름으로 복사
# cp -r [원본디렉토리] [대상디렉토리]: 디렉토리 전체를 복사

!cp
```

```python
# MoVe 의 mv
# 파일이나 디렉토리를 이동시키거나, 이름을 변경할 때 사용
# 리눅스에서는 별도의 '이름 바꾸기' 명령어가 따로 없고, mv를 사용하여 같은 위치에서 이름만 바꾸는 방식을 씁니다.

# mv old_name new_name: 파일 이름 변경
# mv file.txt /home/user/: 파일을 해당 경로로 이동

!mv
```

```python
# MaKe DIRectory 의 mkdir
# 새로운 디렉토리(폴더)를 생성

# mkdir [폴더명]: 폴더 생성
# mkdir -p path/to/dir: 하위 폴더까지 한 번에 생성 (Parents)

!mkdir
```

```python
# conCATenate (연결시키다)
#파일의 내용을 터미널 화면에 출력하거나, 여러 파일을 하나로 합침

# cat [파일명]: 파일의 전체 내용을 화면에 출력
# cat -n [파일명]: 각 줄 앞에 행 번호를 붙여서 출력 (Number)
# cat [파일A] [파일B] > [파일C]: 파일 A와 B의 내용을 합쳐서 새로운 파일 C 생성

# 파일 내용이 너무 길면 터미널 화면을 순식간에 지나가 버려 확인하기 어려움.
# 이럴 때는 한 페이지씩 끊어서 보여주는 less 명령어를 함께 사용

!cat
```

```python
# FILE (파일)
# 파일의 '실제 형식(Type)' 이 무엇인지 확인합니다.

# file [파일명]: 파일의 종류를 확인 (가장 기본)
# file -b [파일명]: 파일 이름은 빼고 결과만 간략하게 출력 (Brief)
# file -i [파일명]: MIME 타입 형식으로 출력 (예: text/plain, image/png)

# 리눅스에서는 파일 확장자(.txt, .jpg 등)가 절대적인 기준이 아님
# 확장자가 없거나 잘못 붙어 있어도 file 명령어를 사용하면 파일 내부의 '매직 넘버(Header)'를 읽어 실제 정체를 정확히 알려줌

!file
```

```python
# HISTORY (기록/이력)

# 터미널에서 이전에 입력했던 명령어들의 목록을 순서대로 보여줍니다.

# history: 입력했던 모든 명령어 목록을 번호와 함께 출력

# history [n]: 최근 입력한 n개의 명령어만 확인 (예: history 10)

# !n: 목록에 표시된 n번 명령어를 즉시 재실행 (예: !105)

# !!: 바로 직전에 입력했던 명령어를 다시 실행

# history -c: 현재 세션의 명령어 기록을 모두 삭제 (Clear)

# 길고 복잡한 명령어를 다시 입력해야 할 때 유용
# 특히 history 명령어로 번호를 확인한 뒤 ! + 번호 를 입력하면 오타 없이 빠르게 작업할 수 있음

!history
```

```python
# CHange MODe
# 파일이나 디렉토리의 **권한(읽기, 쓰기, 실행)**을 변경

# chmod 755 [파일명]: 소유자는 모든 권한(7), 그룹과 나머지는 읽기/실행(5) 권한만 부여
# chmod +x [파일명]: 해당 파일에 실행(eXecute) 권한을 추가 (스크립트 실행 시 필수)
# chmod -R 644 [디렉토리]: 하위 폴더와 파일의 권한을 모두 한꺼번에 변경 (Recursive)

# 권한은 숫자(R=4, W=2, X=1)의 합으로 계산 (예: 7(4+2+1, 모든 권한), 5(4+1, 읽기/실행))

!chmod
```

```python
# CHange OWNer

# 파일이나 디렉토리의 소유자 또는 소유 그룹을 변경

# chown [사용자명] [파일명]: 파일의 소유자를 해당 사용자로 변경
# chown [사용자]:[그룹명] [파일명]: 소유자와 그룹을 동시에 변경
# chown -R [사용자] [디렉토리]: 디렉토리 내부의 모든 파일 소유자를 한꺼번에 변경

# 시스템의 중요한 소유권을 변경하는 작업이므로, 대부분의 경우 명령어 앞에 "sudo" 를 붙여서 실행

!chown
```
