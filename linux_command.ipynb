```python
# chmod

# CHange MODe
# 파일이나 디렉토리의 **권한(읽기, 쓰기, 실행)**을 변경

# chmod 755 [파일명]: 소유자는 모든 권한(7), 그룹과 나머지는 읽기/실행(5) 권한만 부여
# chmod +x [파일명]: 해당 파일에 실행(eXecute) 권한을 추가 (스크립트 실행 시 필수)
# chmod -R 644 [디렉토리]: 하위 폴더와 파일의 권한을 모두 한꺼번에 변경 (Recursive)

# 권한은 숫자(R=4, W=2, X=1)의 합으로 계산 (예: 7(4+2+1, 모든 권한), 5(4+1, 읽기/실행))
```

```python
# chown

# CHange OWNer
# 파일이나 디렉토리의 소유자 또는 소유 그룹을 변경

# chown [사용자명] [파일명]: 파일의 소유자를 해당 사용자로 변경
# chown [사용자]:[그룹명] [파일명]: 소유자와 그룹을 동시에 변경
# chown -R [사용자] [디렉토리]: 디렉토리 내부의 모든 파일 소유자를 한꺼번에 변경

# 시스템의 중요한 소유권을 변경하는 작업이므로, 대부분의 경우 명령어 앞에 "sudo" 를 붙여서 실행
```

```python
# ln

# LiNk (연결)
# 특정 파일이나 디렉토리에 연결된 "링크(Link)" 를 생성. 윈도우의 '바로가기 아이콘'을 만드는 것과 유사

# ln -s [대상] [링크이름]: 심볼릭 링크(Symbolic Link) 생성. 가장 많이 사용되며, 원본의 경로를 가리키는 소프트 링크
#                          (예: ln -s /usr/bin/python3 python)
# ln [대상] [링크이름]: 하드 링크(Hard Link) 생성. 원본 파일과 동일한 데이터를 가리키는 복제본 같은 링크
#                       (원본이 지워져도 데이터가 유지됨)

# 심볼릭 링크(-s)를 만든 경우, 원본 파일을 삭제하거나 위치를 옮기면 링크가 깨져서 더 이상 사용할 수 없음


# 심볼릭 링크 (Soft Link, -s 옵션):
    # 가장 대중적
    # 파일뿐만 아니라 디렉토리에도 링크를 걸 수 있음
    # 원본 파일의 '주소'만 가지고 있는 포스트잇 같은 존재
# 하드 링크 (Hard Link):
    # 원본 파일의 실제 데이터 위치(Inode)를 직접 공유
    # 원본 파일이 삭제되어도 하드 링크 파일이 남아있다면 데이터를 읽을 수 있음
    # 디렉토리에는 생성할 수 없다는 제약이 있음
```

```python
# apt

# Advanced Package Tool (고급 패키지 도구)
# 우분투에서 소프트웨어를 설치, 업데이트, 업그레이드, 삭제하는 패키지 관리 도구

# sudo apt update: 설치 가능한 패키지들의 최신 목록을 가져옵니다. (설치 전 필수!)
# sudo apt upgrade: 설치된 모든 프로그램을 최신 버전으로 업데이트합니다.
# sudo apt install [패키지명]: 새로운 프로그램을 설치합니다.
# sudo apt remove [패키지명]: 프로그램을 삭제합니다.
# apt search [키워드]: 설치할 수 있는 패키지가 있는지 검색합니다. (이건 sudo가 없어도 됩니다.)

# 시스템의 중요한 소프트웨어를 건드리는 작업이 많으므로,
# 검색(search)을 제외한 대부분의 명령어 앞에 "sudo" 를 꼭 붙여야 함.
```

```python
# snap

# SNAP (딱 들어맞다, 낚아채다)
# 앱과 그 앱이 실행되는 데 필요한 모든 라이브러리/의존성을 하나로 묶어 제공하는 패키지 관리 도구
# 시스템 환경에 구애받지 않고 어디서든 동일하게 작동하도록 설계

# sudo snap install [패키지명]: 프로그램 설치
# sudo snap remove [패키지명]: 프로그램 삭제
# snap list: 설치된 snap 패키지 목록 확인
# snap find [키워드]: 설치 가능한 패키지 검색 (예: snap find vscode)
# sudo snap refresh: 모든 snap 패키지를 최신 버전으로 업데이트

# apt보다 설치 용량이 크고, 첫 실행 속도가 약간 느릴 수 있음.
# 보안을 위해 '샌드박스'라는 격리된 공간에서 실행되므로 시스템의 특정 파일에 접근할 때 권한 문제가 발생할 수 있음

# 구분   |                 apt                 |                       snap
# 의존성 | 	시스템에 설치된 라이브러리 공유 	 |     필요한 모든 것을 패키지 안에 포함
# 버전	 |   배포판(우분투 버전)에 따라 다름	 |     어떤 배포판이든 항상 최신 버전 유지 가능
# 보안	 |   시스템 권한을 가짐	               |     "격리된 공간(Sandbox)"에서 실행
```

```python
# tar

# Tape ARchiver (테이프 보관기)
# 여러 개의 파일이나 디렉토리를 하나의 파일로 묶거나(Archive), 묶인 파일을 다시 푼다.
# (자체 압축 기능보다는 '묶기'가 주 목적이지만, 옵션을 통해 압축도 동시에 수행)

# tar -cvf [파일명.tar] [대상]: 대상들을 하나로 묶기 (Create)
# tar -xvf [파일명.tar]: 묶인 파일을 풀기 (eXtract)
# tar -zvcf [파일명.tar.gz] [대상]: 묶으면서 동시에 압축 (Gzip 사용)
# tar -zvxf [파일명.tar.gz]: 압축된 파일을 동시에 해제

# -f 옵션은 바로 뒤에 파일명이 와야 하므로 보통 명령어의 가장 끝에 붙여서 사용
    # (예: -cvzf는 가능하지만 -cfvz는 오류가 날 수 있음)

# 💡 옵션 글자 한눈에 이해하기
# c (Create): 새로운 아카이브 파일을 생성
# x (eXtract): 아카이브 파일의 내용을 추출하여 풀어냄.
# v (Verbose): 진행 과정을 화면에 상세히 보여줌
# f (File): 작업할 아카이브 파일의 이름을 지정 (필수)
# z (Gzip): .gz 형식으로 압축하거나 해제
```

```python
# find

# FIND (찾다)
# 파일 시스템에서 다양한 조건(이름, 크기, 시간, 종류 등)을 기준으로 파일이나 디렉토리를 검색

# find [경로] -name "[파일명]": 파일 이름으로 검색 (예: find . -name "*.java")
# find [경로] -type [f/d]: 파일(f)만 찾을지, 디렉토리(d)만 찾을지 지정
# find [경로] -size +[크기]: 특정 크기 이상의 파일을 검색 (예: +100M)
# find [경로] -empty: 내용이 비어 있는 파일이나 폴더 검색
# find [경로] -exec [명령어] {} \;: 검색된 결과에 대해 특정 명령어를 실행 (예: 찾은 파일을 모두 삭제)

# 경로를 루트(/)로 설정하고 검색하면 시스템 전체 파일을 다 뒤져야 하므로 시간이 매우 오래 걸릴 수 있으니까
# 가급적 특정 하위 폴더에서 검색하는 것이 효율적
```

```python
# systemctl

# SYSTEM ConTroL (시스템 제어)
# 리눅스의 시스템 및 서비스 매니저인 systemd를 제어
# 프로그램(서비스)을 백그라운드에서 실행하거나, 부팅 시 자동으로 켜지게 설정하는 등 시스템 전반을 관리

# sudo systemctl start [서비스명]: 서비스 시작
# sudo systemctl stop [서비스명]: 서비스 중지
# sudo systemctl restart [서비스명]: 서비스 재시작 (설정 변경 후 적용할 때 자주 사용)
# systemctl status [서비스명]: 서비스의 현재 상태 확인 (정상 작동 여부, 최근 로그 등)
# sudo systemctl enable [서비스명]: 부팅 시 자동 실행되도록 설정
# sudo systemctl disable [서비스명]: 부팅 시 자동 실행 해제

# 시스템의 서비스를 직접 건드리는 작업이므로 거의 모든 명령어 앞에 sudo 권한이 필요
```

```python
# ifconfig

# InterFace Configuration (인터페이스 설정)
# 네트워크 인터페이스(랜카드 등)의 상태를 확인하거나, IP 주소 및 서브넷 마스크 등을 설정합니다.

# ifconfig: 활성화된 모든 네트워크 인터페이스의 정보(IP, MAC 주소 등)를 출력
# ifconfig -a: 비활성화된 것을 포함한 모든 인터페이스 확인
# sudo ifconfig [인터페이스명] up/down: 특정 인터페이스를 활성화하거나 비활성화
# sudo ifconfig [인터페이스명] [IP주소]: 특정 인터페이스에 IP 주소를 수동으로 할당

# 최신 리눅스 배포판(우분투 20.04 이후 등)에서는 ifconfig가 포함된 net-tools 패키지가 기본으로 설치되지 않는 경우가 많음.
# 만약 명령어가 없다면 sudo apt install net-tools로 설치하거나, 최신 표준인 ip addr 명령어를 사용해야 합니다.
```

```python
# nmtui

# Network Manager Text User Interface (네트워크 관리자 텍스트 사용자 인터페이스)
# 터미널 환경에서 "그래픽 메뉴(TUI)"를 띄워 네트워크 연결
# IP 설정, 호스트 이름 변경 등을 마우스 없이 화살표 키로 편리하게 설정

# nmtui: 전체 메뉴 실행 (연결 편집, 연결 활성화, 호스트 이름 설정)
# nmtui edit: 특정 네트워크 인터페이스 편집 화면으로 바로 진입
# nmtui connect: 와이파이(Wi-Fi)나 유선 네트워크 연결 선택 화면으로 진입

# 복잡한 명령어나 설정 파일(netplan 등)을 직접 수정하기 어려울 때 매우 유용
# 서버 환경(CLI 전용)에서 네트워크를 잡을 때 '치트키' 같은 도구
```

```python
# nslookup

# Name Server Lookup (네임 서버 조회)
# DNS(도메인 네임 시스템) 서버에 질의하여 특정 도메인의 IP 주소를 찾아내거나,
# 반대로 IP 주소로 도메인 이름을 확인

# nslookup [도메인명]: 해당 도메인의 IP 주소 확인 (예: nslookup google.com)
# nslookup [IP주소]: 해당 IP에 연결된 도메인 확인 (역방향 조회)
# nslookup [도메인명] [DNS서버주소]: 특정 DNS 서버(예: 구글 8.8.8.8)를 지정하여 조회

# 결과에 'Non-authoritative answer'라고 나오는 것은 캐시된 정보를 가져왔다는 뜻
# 정보 자체가 틀렸다는 의미는 아니다.

```

```python
# ping

# Packet InterNet Groper (패킷 인터넷 탐색기)
# ICMP(Internet Control Message Protocol)를 사용
# 특정 대상(IP 또는 도메인)에 신호를 보낸 뒤,
# 응답이 돌아오는지를 확인하여 네트워크 연결 유무와 지연 시간(Latency)을 측정

# ping [주소]: 중단할 때까지 무한히 패킷을 전송 (Ctrl + C로 종료)
# ping -c [횟수] [주소]: 지정한 횟수만큼만 패킷을 보내고 통계를 보여준 뒤 종료 (Count)
# ping -i [초] [주소]: 패킷을 보내는 시간 간격을 설정 (Interval)
# ping -s [바이트] [주소]: 전송하는 패킷의 크기를 변경. 네트워크 부하 테스트 시 활용 (Size)

# 보안을 위해 서버의 방화벽에서 ICMP 응답을 차단한 경우, 서버가 실제로 켜져 있어도 "Request timeout"이 발생할 수 있음.
```

```python
# ps
# Process Status (프로세스 상태)
# 현재 시스템에서 실행 중인 프로세스(프로그램)들의 목록과 상태를 보여줍니다.

# ps: 현재 터미널 창(세션)에서 실행 중인 프로세스만 간단히 출력
# ps -e: 시스템에서 실행 중인 모든(Every) 프로세스를 출력
# ps -f: 프로세스의 상세 정보(사용자, PID, 시작 시간 등)를 포함하여 출력 (Full format)
# ps -ef: 모든 프로세스를 상세한 정보와 함께 리스트 형태로 출력 (가장 표준적인 사용법)
# ps aux: 실행 중인 모든 프로세스를 CPU/메모리 사용량과 함께 출력 (BSD 스타일 방식)

# ps는 명령어를 입력한 그 순간의 스냅샷만 보여준다.
# 만약 실시간으로 변하는 상태를 보고 싶다면 top이나 htop 명령어를 사용

# ps -ef나 ps aux를 입력했을 때 나오는 항목 중 꼭 알아야 할 것들
#   항목       |       의미	     |                 설명
# UID/USER     | 	사용자	       |    프로세스를 실행한 권한을 가진 사용자 이름
# PID	         |   프로세스 ID   |   	프로세스마다 부여된 고유 번호 (kill 할 때 사용)
# PPID	       |   부모 PID      | 	  이 프로세스를 만든 '부모' 프로세스의 번호
# STIME/START	 |   시작 시간	   |    프로세스가 시작된 시각
# CMD/COMMAND	 |   실행 명령어	 |    프로세스를 실행한 명령어의 경로와 이름
```

```python
# grep
# Global Regular Expression Print (전역 정규 표현식 출력)
# 파일 내부나 터미널 출력 결과 중에서 특정 문자열이나 패턴을 포함한 행을 찾아 화면에 보여줍니다.

# grep "[문자열]" [파일명]: 파일에서 특정 단어가 포함된 줄 찾기
# grep -i "[문자열]" [파일명]: 대소문자를 구분하지 않고 검색 (Ignore Case)
# grep -r "[문자열]" [디렉토리]: 하위 폴더의 모든 파일을 뒤져서 검색 (Recursive)
# grep -v "[문자열]" [파일명]: 해당 단어가 포함되지 않은 줄만 보여주기 (Invert)
# grep -n "[문자열]" [파일명]: 검색 결과가 몇 번째 줄에 있는지 번호 표시 (Line Number)

# 다른 명령어의 결과물과 함께 쓰는 경우가 많다.
    # 예: history | grep "sudo" (입력했던 명령어 중 sudo가 들어간 것만 필터링)
```

```python
# kill
# KILL (죽이다 / 종료시키다)
# 실행 중인 프로세스에 신호(Signal)를 보내서 종료하거나 상태를 변경
# 단순히 '끄는' 것뿐만 아니라, 다양한 방식으로 프로세스를 제어

# kill [PID]: 해당 프로세스 ID(PID)에 종료 신호를 보냄 (기본값: SIGTERM)
# kill -9 [PID]: 프로세스를 강제 종료 (SIGKILL, 어떤 상황에서도 즉시 종료)
# kill -15 [PID]: 프로세스가 하던 일을 정리하고 안전하게 종료되도록 유도 (권장 방식)
# kill -l: 보낼 수 있는 모든 신호(Signal)의 종류를 확인

# kill 명령어를 쓰려면 먼저 종료할 프로그램의 "PID(프로세스 번호)"를 알아야 한다.
    # (ps나 top 명령어로 확인 가능)

# 번호  |       이름      |              설명
# 15	  |      SIGTERM	  |       정상 종료. 메모리를 정리하고 안전하게 닫음 (기본값)
# 9	    |      SIGKILL	  |       강제 종료. 어떤 작업 중이라도 즉시 프로세스를 소멸
# 1	    |      SIGHUP     |       재시작. 설정 파일을 다시 읽어오게 할 때 사용
# 19	  |      SIGSTOP	  |       프로세스를 일시 정지
```

```python
# | (shift + \)
# Pipe (파이프)
# 앞 명령어의 **출력(결과물)**을 뒤 명령어의 입력으로 바로 전달
# 두 명령어를 연결하는 통로 역할

# [명령어1] | [명령어2]: 명령어1의 결과에서 명령어2를 실행 (예: ls -al | grep "test")

# grep, wc, awk 등과 함께 쓰여 방대한 데이터에서 필요한 것만 필터링할 때 필수적으로 사용
```

```python
# >
# Redirection (리다이렉션 / 방향 재지정)
# 화면에 출력될 결과를 화면이 아닌 파일로 저장하거나 방향을 바꿈

# [명령어] > [파일명]: 결과를 파일로 저장 (덮어쓰기)
# [명령어] >> [파일명]: 결과를 파일 끝에 추가 (이어쓰기)
# [명령어] < [파일명]: 파일의 내용을 명령어의 입력으로 사용

# > 기호는 기존 파일 내용을 지우고 새로 쓰기 때문에, 중요한 파일이 있는 곳에서 사용할 때 주의
```

```python
# &
# Background (백그라운드 실행)
# 명령어를 뒷단(백그라운드)에서 실행하게 합니다. 명령어가 끝나기를 기다리지 않고 바로 다음 명령어를 입력 가능

# [명령어] &: 해당 명령어를 백그라운드로 돌리고 터미널 프롬프트를 즉시 반환

# 실행 시간이 오래 걸리는 작업(대용량 압축, 서버 실행 등)을 실행해두고 다른 작업을 계속하고 싶을 때 사용
```

```python
# fdisk
# Fixed Disk (고정 디스크)
# 디스크의 **파티션(Partition)**을 생성, 삭제, 수정하는 도구
# 커다란 피자(디스크)를 조각(파티션)으로 나누는 작업

# sudo fdisk -l: 시스템에 연결된 모든 디스크와 파티션 목록 확인
# sudo fdisk /dev/sdb: 특정 디스크(sdb)의 파티션 설정 모드 진입

# 파티션 테이블을 직접 건드리는 작업이므로, 실수로 설정을 저장하면 기존 데이터가 모두 날아갈 수 있다.
```

```python
# mkfs.ext4
# Make File System (ext4)
# 나눈 파티션을 **포맷(Format)**하여 데이터를 담을 수 있는 상태로 변경.
# 나눈 피자 조각 위에 어떤 토핑을 올릴지(어떤 파일 시스템을 쓸지) 정하는 단계

# sudo mkfs.ext4 /dev/sdb1: sdb1 파티션을 리눅스 표준인 ext4 형식으로 포맷

# 실행 즉시 해당 파티션의 모든 내용이 삭제
# 명령어 뒤에 파티션 번호(숫자)를 정확히 확인하고 입력
```

```python
# df
# Disk Free (디스크 여유 공간)
# 시스템에 마운트된 모든 파일 시스템의 전체 용량, 사용량, 남은 용량 등을 보여줌

# df: 기본 단위(1KB)로 디스크 정보를 출력 (숫자가 너무 커서 읽기 힘들다.)
# df -h: 용량을 사람이 읽기 편하게 GB, MB 단위로 표시 (Human-readable, 가장 많이 사용함)
# df -T: 파일 시스템의 종류(ext4, xfs, tmpfs 등)를 함께 보여줌 (Type)
# df -i: 용량 대신 아이노드(Inode) 사용 정보를 보여줌 (파일 개수가 너무 많아 공간이 부족할 때 확인)

# df는 파일 시스템 전체의 상태를 보여줌
# 특정 폴더가 얼마나 차지하는지 알고 싶다면 du (Disk Usage) 명령어를 사용
```

```python
# mount
# Mount (장착하다 / 올라타다)
# 포맷된 파티션을 리눅스의 특정 디렉토리(폴더)와 연결
# 리눅스에서는 장치를 폴더처럼 열어서 사용하기 때문에 이 과정이 필수

# sudo mount [장치명] [폴더경로]: 장치를 특정 폴더에 연결 (예: sudo mount /dev/sdb1 /mnt/data)
# mount: 현재 마운트된 모든 장치의 목록 확인

# 연결할 폴더(마운트 포인트)는 미리 만들어져 있어야 함.
```

```python
# umount
# Unmount (해제하다)
# 연결된 파티션의 마운트를 해제
# USB를 뽑기 전에 윈도우에서 '안전하게 제거'를 누르는 것과 같다.

# sudo umount [장치명] 또는 sudo umount [폴더경로]: 마운트 해제

# 현재 그 폴더 안에서 작업을 하고 있거나(cd로 들어가 있는 경우 등),
# 파일이 열려 있으면 "target is busy"라는 메시지와 함께 해제가 되지 않는다.
```

```python
# echo
# Echo (메아리)
# 입력한 문자열을 터미널에 출력하거나, 시스템에 설정된 환경 변수의 값을 확인할 때 사용

# echo "[문자열]": 입력한 문자열을 그대로 화면에 출력
# echo $[변수명]: 특정 환경 변수의 값을 출력 (예: echo $PATH)
# echo -e "[문자열]": 줄바꿈(\n)이나 탭(\t) 같은 특수 문자를 해석해서 출력
# echo "[문자열]" > [파일명]: 출력 내용을 화면이 아닌 파일로 저장 (Redirection)

# 큰따옴표(")와 작은따옴표(')의 차이가 있다.
# 큰따옴표는 그 안의 변수(예: $USER)를 해석해서 출력
# 작은따옴표는 기호 그대로를 문자로 취급
```

```python
# export
# Export (내보내기)
# 현재 쉘에서 정의한 변수를 **환경 변수(Environment Variable)**로 격상
# 이후에 실행되는 모든 자식 프로세스(프로그램)들이 이 변수를 사용할 수 있게 함

# export [변수명]=[값]: 환경 변수 설정 (예: export MY_NAME=Gemini)
# export [변수명]=$[변수명]:[추가값]: 기존 변수 값에 새로운 경로 등을 추가 (예: export PATH=$PATH:/home/user/bin)

# export로 설정한 변수는 현재 터미널 세션이 종료되면 사라짐
# 영구적으로 적용하려면 ~/.bashrc나 ~/.profile 파일에 기록
```

```python
# printenv
# Print Environment (환경 출력)
# 현재 시스템에 설정된 모든 환경 변수를 화면에 출력
# 특정 변수 이름만 지정해서 해당 값만 볼 수도 있

# printenv: 모든 환경 변수 목록 확인
# printenv [변수명]: 특정 환경 변수의 값만 확인 (예: printenv SHELL)

# echo $[변수명]과 비슷하지만, printenv는 인자 없이 실행했을 때 전체 목록을 보여준다는 점이 다르다.
```

```python
# telnet
# Teletype Network (텔레타이프 네트워크)
# 원격지의 컴퓨터에 접속하기 위한 프로토콜...
# 현재는 보안상의 이유로(데이터 암호화 안 됨) SSH에 자리를 빼앗김.
# 지금은 주로 특정 서버의 포트가 열려 있는지 확인하는 용도로 사용

# telnet [IP주소/도메인] [포트번호]: 해당 서버의 포트가 열려 있는지 확인 (예: telnet google.com 80)
# 종료 방법: 접속 성공 시 Ctrl + ]를 누른 후 quit 입력

# 최신 우분투에는 기본 설치되어 있지 않은 경우가 많다.
# 사용하려면 sudo apt install telnet으로 설치
```

```python
# ssh
# Secure Shell (보안 쉘)
# 네트워크 상의 다른 컴퓨터에 안전하게 접속하거나 원격 명령을 실행
# 모든 통신 데이터가 암호화되어 전송되기 때문에
# 아이디, 비밀번호와 같은 민감한 정보가 중간에 탈취되지 않도록 보호

# ssh [사용자명]@[IP주소]: 원격 서버에 기본 포트(22번)로 접속
# ssh -p [포트번호] [사용자명]@[IP주소]: 기본 포트가 아닌 다른 포트로 접속 (보안상 포트를 변경한 경우 사용)
# ssh [사용자명]@[IP주소] "[명령어]": 서버에 직접 로그인하지 않고, 특정 명령어 하나만 원격으로 실행하고 즉시 종료
# exit: 현재 연결된 SSH 세션을 종료하고 내 컴퓨터로 돌아옴

# 처음 접속하는 서버는 "이 서버를 신뢰하느냐"는 확인 메시지(Fingerprint)가 뜬다. 이때 yes를 입력해야 접속이 가능
# 또한, 보안을 위해 매번 비밀번호를 치는 대신 **SSH 키(Key)**를 등록하여 접속하는 방식이 권장
```

```python
# ufw
# Uncomplicated FireWall (복잡하지 않은 방화벽)
# 리눅스 커널의 패킷 필터링 기능(iptables)을 훨씬 쉽게 다룰 수 있게 만든 방화벽 관리 도구
# 외부의 불법적인 접근을 막거나 특정 포트만 열어주는 보안 역할

# sudo ufw status verbose: 방화벽의 현재 작동 상태와 적용된 규칙들을 상세히 확인
# sudo ufw enable / disable: 방화벽을 활성화하거나 비활성화
# sudo ufw allow [포트번호/서비스]: 특정 포트나 서비스의 접속을 허용 (예: allow 8080, allow ssh)
# sudo ufw deny [포트번호/서비스]: 특정 접속을 차단
# sudo ufw delete allow [포트번호]: 기존에 설정한 허용 규칙을 삭제

# 원격(SSH)으로 접속 중인 서버에서 sudo ufw enable을 할 때는 반드시 먼저 sudo ufw allow ssh를 실행
# 그렇지 않으면 방화벽이 켜지는 순간 자신의 접속도 차단되어 서버에 다시 들어갈 수 없는 상황이 발생
```

```python
#xrdp (X Remote Desktop Protocol)
# X (리눅스 그래픽 시스템) + RDP (Microsoft 원격 데스크톱 프로토콜)
# 윈도우의 '원격 데스크톱 연결(mstsc)' 기능을 이용해 리눅스 데스크탑 환경에 접속할 수 있게 해주는 오픈소스 서버 소프트웨어입니다.

# 설치: sudo apt install xrdp
# 상태 확인: sudo systemctl status xrdp
# 방화벽 허용: sudo ufw allow 3389/tcp (RDP 기본 포트인 3389번을 열어야 합니다.)
# 권한 설정: sudo adduser xrdp ssl-cert (인증서 접근 권한 부여)

# xrdp를 통해 접속할 때는 리눅스 로컬(직접 컴퓨터 앞)에서 이미 로그인이 되어 있으면 안된다.
# 세션이 충돌하여 접속하자마자 튕기거나 검은 화면이 나올 수 있으니, 반드시 로컬 로그아웃을 먼저 진행
```

```python
# VNC
# Virtual Network Computing (가상 네트워크 컴퓨팅)
# RFB(Remote Frame Buffer) 프로토콜을 사용
# 네트워크를 통해 멀리 떨어진 컴퓨터의 그래픽 화면(GUI)을 실시간으로 공유하고 제어
# VNC Server: 화면을 제공하는 쪽 (예: TigerVNC, TightVNC, RealVNC 등)
# VNC Viewer (Client): 접속해서 조작하는 쪽

# vncserver: VNC 서버를 시작하고 가상 데스크톱 생성 (비밀번호 설정 포함)
# vncserver -list: 현재 열려 있는 VNC 세션 목록 확인
# vncserver -kill :[번호]: 특정 가상 데스크톱 종료 (예: vncserver -kill :1)

# VNC 자체는 데이터를 암호화하지 않고 전송하기 때문에 보안에 취약
# 외부망에서 접속할 때는 반드시 SSH 터널링을 함께 사용하는 것이 표준 보안 수칙

# 🔍 xrdp vs VNC, 어떤 걸 써야 할까?
# 구분	            |                 xrdp	                |                    VNC
# 접속 클라이언트	  |   윈도우 기본 '원격 데스크톱' (RDP)	  |      별도의 VNC 뷰어 프로그램 필요
# 속도/성능	        |   비교적 부드러움 (최적화 우수)	      |      네트워크 환경에 따라 느릴 수 있음
# 보안	            |   기본 암호화 지원	                  |      암호화 없음 (SSH 터널링 권장)
# 편의성            | 	윈도우 사용자에게 매우 친숙함	      |      리눅스 간 혹은 이기종 간 범용성 높음
```
